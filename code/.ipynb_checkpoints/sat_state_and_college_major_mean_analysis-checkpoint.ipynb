# Background

The SAT is standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). For information on SAT can be found on their website:
- [SAT](https://collegereadiness.collegeboard.org/sat)

For SAT test results([_source_](https://collegereadiness.collegeboard.org/sat/scores/understanding-scores/interpreting)), we know the following
- Total = Evidence-Based Reading and Writing + Math
- 200 <= Evidence-Based Reading and Writing <= 800
- 200 <= Math <= 800
- 400 <= Total <= 1600

_*For the below writeup and codes, we will use EBRW to represent Evidence-Based Reading and Writing_

1. Manually calculate mean:

    Write a function that takes in values and returns the mean of the values. Create a list of numbers that you test on your function to check to make sure your function works!
    
    *Note*: Do not use any mean methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# function to find mean of values without any Python libraries
def find_mean(values):
    return (sum(values) / len(values))
```

```python
#import numpy for verification of function
import numpy as np

# verify function against python library
print(f'Mean using function for [1,2,3,4,5]: {find_mean([1,2,3,4,5])}')
print(f'Mean using numpy for [1,2,3,4,5]: {np.mean([1,2,3,4,5])}')
print(f'---')
print(f'Mean using function for range(100): {find_mean(range(100))}')
print(f'Mean using numpy for range(100): {np.mean(range(100))}')
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# function to find standard deviation without any Python libraries
def find_sd(x_values):
    n = len(x_values)
    miu = find_mean(x_values)
    
    # Summation of (x - miu)**2
    summation = 0               # counter
    for x in x_values:
        summation += (x - miu)**2
        
    sd = ((1/n)*summation)**(1/2)    #calculate sd
    
    return sd
```

#### Comment on standard deviation function
Standard Deviation calculated using the above function will be the same as the `numpy` standard deviation but it will be different from the value calculated using `pandas`. Updating the function to `n - 1` instead of `n` will result in same value as the pandas in `line 11`. 

Setting `ddof=1` in `numpy` standard deviation will get the same result as `pandas` function. 

For more information, refer to link:
https://stackoverflow.com/questions/24984178/different-std-in-pandas-vs-numpy

```python
# verify function against python library
print(f'SD using function for [1,2,3,4,5]: {find_sd([1,2,3,4,5])}')
print(f'SD using numpy for [1,2,3,4,5]: {np.std([1,2,3,4,5])}')
print(f'---')
print(f'SD using function for range(100): {find_sd(range(100))}')
print(f'SD using numpy for range(100): {np.std(range(100))}')
```

3. Data cleaning function:
    
    Write a function that takes in a string that is a number and a percent symbol (ex. '50%', '30.5%', etc.) and converts this to a float that is the decimal approximation of the percent. For example, inputting '50%' in your function should return 0.5, '30.5%' should return 0.305, etc. Make sure to test your function to make sure it works!

You will use these functions later on in the project!

```python
# function to change percent to float
# e.g. 50% -> 0.5
def percent_to_float(string):
    new_percent = np.nan     # returns null value if there is no % defined
    # new_percent = None
    if '%' in string:
        new_num = float(string.replace('%',''))
        new_percent = new_num/100
    return new_percent
```

```python
# verify function is working
print(percent_to_float('50%'))
print(percent_to_float('110%'))
print(percent_to_float('0.5%'))
```

Additional function created for data cleaning

```python
# function to convert numbers with comma
def comma_to_int(string):
    if ',' in string:
        new_num = int(string.replace(',',''))
    else:
        new_num = int(string)     # returns null value if there is no , within
    return new_num
```

```python
# verify function is working
print(comma_to_int('1,234'))
print(comma_to_int('1,111,333'))
```

### Count of State SAT Mean Above College Major SAT Mean

We will create a dataframe comparing state average score against college major average score.

Looking back at the problem statement, we want to boost qualification rate, and with a higher average than the college major average, it will likely increase the chances of the getting accepted into the major. With this, we will be comparing and getting a count of values and storing it in a dataframe for analysis

```python
# Using list comprehension, create dictionary containing 
# the states and the count of scores above college average
count_sat_dict = {
    'state': [state for state in df_sat_combine['state']],
    '2019_count_total': [(df_sat_college_combine['2019_college_total'] < total).sum()
                                      for total in df_sat_combine[['2019_total']].itertuples(index=False)],
    '2019_count_ebrw': [(df_sat_college_combine['2019_college_ebrw'] < ebrw).sum()
                                      for ebrw in df_sat_combine[['2019_ebrw']].itertuples(index=False)],
    '2019_count_math': [(df_sat_college_combine['2019_college_math'] < math).sum()
                                      for math in df_sat_combine[['2019_math']].itertuples(index=False)],
    '2018_count_total': [(df_sat_college_combine['2018_college_total'] < total).sum()
                                      for total in df_sat_combine[['2018_total']].itertuples(index=False)],
    '2018_count_ebrw': [(df_sat_college_combine['2018_college_ebrw'] < ebrw).sum()
                                      for ebrw in df_sat_combine[['2018_ebrw']].itertuples(index=False)],
    '2018_count_math': [(df_sat_college_combine['2018_college_math'] < math).sum()
                                      for math in df_sat_combine[['2018_math']].itertuples(index=False)]
                 }

#create dataframe
df_sat_state_count_mean = pd.DataFrame(count_sat_dict)
```

## Heatmap
Lets take a look at the columns and see if there is any relationship between them for the 3 dataframes.

We will use a function to create heatmaps for the 2 dataframes.

```python
# function for graph plotting
# dataset1 : dataframe
# dataset2 : dataframe
# title : string - graph title
# labels : list - subplot titles
# color : string - color style

def heatmap_plot(dataset1, dataset2, title, labels, color):
    #creating mask for the heatmap
    mask1 = np.zeros_like(dataset1.corr())
    mask1[np.triu_indices_from(mask1)] = True
    
    mask2 = np.zeros_like(dataset2.corr())
    mask2[np.triu_indices_from(mask2)] = True

    # heatmap on college major SAT score mean
    fig, ax = plt.subplots(1, 2, figsize=(12,6))
    fig.suptitle(title)
    
    sns.heatmap(ax=ax[0], data=dataset1.corr(), annot=True, linewidth = 0.3, mask=mask1, cmap=color)
    ax[0].set_title(labels[0])
    
    sns.heatmap(ax=ax[1], data=dataset2.corr(), annot=True, linewidth = 0.3, mask=mask2, cmap=color)
    ax[1].set_title(labels[1])
    
    plt.tight_layout()    #help to add more space/adjust the plots to best layout (like a print job, best fit)
    
    plt.show()
```

## Histograms
Lets take a look at histograms to observe the distribution of the data. As we will be plotting similar histograms for comparison, we will use a function for the graph plotting. 

```python
# Adapted from https://dev.to/thalesbruno/subplotting-with-matplotlib-and-seaborn-5ei8
# Function for histogram plotting
# row : int64 - number of rows
# col: int 64 - number of columns
# xval : list - x-variable for graph plotting
# title : list - graph titles
# dataset : dataframe
# graph_suptitle: figure suptitle
# color: list - graph colors

def histogram_plot(row, col, xval, title, dataset, graph_suptitle, color):
    
    # to have subplots with row and col figsize = (16, 9)
    fig, axes = plt.subplots(row, col, figsize=(16,9), sharey = True)
    fig.suptitle(graph_suptitle)
    
    counter = 0
    
    if (row == 1) or (col == 1):
        for i in range(max(row, col)):
            sns.histplot(ax=axes[i], x = xval[counter], data = dataset, bins = 15, kde=True, color = color[counter])
            axes[i].set_title(title[counter])
            counter += 1
    else:
        for r in range(row):
            for c in range(col):
                sns.histplot(ax=axes[r,c], x = xval[counter], data = dataset, bins = 15, kde=True, 
                             color = color[counter])
                axes[r,c].set_title(title[counter])
                counter += 1

    plt.tight_layout()    #help to add more space/adjust the plots to best layout (like a print job, best fit)

    plt.show()
```

## Boxplot and Violinplot
We will take a look at spread of the distribution and the Inter-Quartile Range (IQR) as well as the 25th Quartile value. For the comparison count, we will be plotting violin plot together with boxplots to show the spread. 

In general, violin plots are a method of plotting numeric data and can be considered a combination of the box plot with a kernel density plot. A violin plot is two KDE plots aligned on an axis. (The kernel density is defined over the full range from -infinity to +infinity.) The “negative” values you are seeing are just an artifact of KDEs. They are estimations of values in your data. It’s not saying we have negative data, it’s saying that this data contains values very close to negative values, namely 0. 


We will use a function for the graphs.

```python
# function for boxplot or violinplot
# graph : list - boxplot or violinplot
# row : int64 - number of rows
# col: int 64 - number of columns
# xval : list - x-variable for graph plotting
# title : list - graph titles
# dataset : dataframe
# graph_suptitle: figure suptitle
# color: list - graph colors

def box_violin_plot(graph, row, col, xval, title, dataset, graph_suptitle, color):
    
    # to have subplots with row and col figsize = (16, 9)
    fig, axes = plt.subplots(row, col, figsize=(16,9), sharex = True)
    fig.suptitle(graph_suptitle)
    
    counter = 0
    
    if (row == 1) or (col == 1):
        for i in range(max(row, col)):
            if graph[counter] == 'boxplot':
                sns.boxplot(ax=axes[i], x = xval[counter], data = dataset, color = color[counter])
                axes[i].axvline(np.percentile(dataset[xval[counter]],25), color='red')
            else:
                sns.violinplot(ax=axes[i], x = xval[counter], data = dataset, color = color[counter])
            axes[i].set_title(title[counter])
            counter += 1
    else: 
        for r in range(row):
            for c in range(col):
                if graph[counter] == 'boxplot':
                    sns.boxplot(ax=axes[r,c], x = xval[counter], data = dataset, color = color[counter])
                    axes[r,c].axvline(np.percentile(dataset[xval[counter]],25), color='red')
                else:
                    sns.violinplot(ax=axes[r,c], x = xval[counter], data = dataset, color = color[counter])
                axes[r,c].set_title(title[counter])
                counter += 1
    
    plt.tight_layout()    #help to add more space/adjust the plots to best layout (like a print job, best fit)

    plt.show()
```

```python
fig, axes = plt.subplots(figsize=(12,6))
fig.suptitle('Graph 1 - High Participation Rate State SAT Mean Total Scores - 2019 and 2018')

# Total scores
sns.barplot(ax=axes, y = 'state', x = 'value',
            data = df_sat_combine.sort_values(
                by='2019_total'
            ).loc[(df_sat_combine['state']=='Colorado') | (df_sat_combine['state']=='Illinois') |
                  (df_sat_combine['state']=='Connecticut') | (df_sat_combine['state']=='Idaho') |
                  (df_sat_combine['state']=='Delaware') | (df_sat_combine['state']=='Florida')|
                  (df_sat_combine['state']=='Michigan') | (df_sat_combine['state']=='Rhode Island'),
                  ['state','2019_total', '2018_total',
                   ]].melt(id_vars=['state'], var_name='Mean_Score'),
           hue='Mean_Score', palette="pastel")

# Mean
axes.axvline(label='2019 State Mean (Total)',
             x = df_sat_combine['2019_total'].mean(),
             color='magenta', linestyle='dashed')
axes.axvline(label='2018 State Mean (Total)',
             x = df_sat_combine['2018_total'].mean(),
             color='blue', linestyle='dotted')

axes.legend(loc="upper left")

plt.tight_layout()    #help to add more space/adjust the plots to best layout (like a print job, best fit)

plt.show()
```
